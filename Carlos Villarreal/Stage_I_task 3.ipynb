Clean up the annual quarter database (i.e. allhlcn20.xlsx) to remove unwanted rows:

```python
import re

# Clean up the annual quarter database
# Filter out rows where "US" appears in the 'St' column
df_quarter_A_filtered = df_quarter_A[~df_quarter_A['St'].str.contains('US', na=False)]

# Filter out rows where "Unknown Or Undefined" appears in the 'Area' column, taking into account case difference
df_quarter_A_filtered = df_quarter_A_filtered[~df_quarter_A_filtered['Area'].str.contains(r'^Unknown or Undefined,\s*\w+', na=False)]
df_quarter_A_filtered = df_quarter_A_filtered[~df_quarter_A_filtered['Area'].str.contains(r'^Unknown or Undefined,\s*\w+', na=False, flags=re.IGNORECASE)]

# Filter out rows where "-- Statewide" appears in the 'Area' column
df_quarter_A_filtered = df_quarter_A_filtered[~df_quarter_A_filtered['Area'].str.contains(r'\b[A-Za-z]+\b\s--\sStatewide', na=False)]

# Filter out rows where there is a NaN or a blank in the 'St' or 'Cnty' column
df_quarter_A_filtered = df_quarter_A_filtered.dropna(subset=['St', 'Cnty'])

print("Annual quarted dataset filtered: \n")
print(df_quarter_A_filtered)

# Save the filtered dataframe as an excel file, comment next line out to not save multiple
#df_quarter_A_filtered.to_excel('filtered_quarter_A.xlsx', index=False)
```

Merge filtered annual quarted database with edited super dataframe that only includes year 2020:

```python
super_team_dataset = pd.read_csv('super_covid19_data_2020.csv')

# Drop rows where countyFIPS is a single 0
super_team_dataset = super_team_dataset[super_team_dataset['countyFIPS'] != 0]

# Standardize countyFIPS to a 5-character string with leading zeros to prevent merging errors
super_team_dataset['countyFIPS'] = super_team_dataset['countyFIPS'].apply(lambda x: f"{int(x):05d}")

# Rename the column in df_quarter_A_filtered to match 'countyFIPS'
df_quarter_A_filtered = df_quarter_A_filtered.rename(columns={'Area\nCode': 'countyFIPS'})

# Exclude redundant columns from super_team_dataset, which are 'County Name_x', 'State', 'County Name_y'
columns_to_exclude = [col for col in super_team_dataset.columns if col not in ['County Name_x', 'State', 'County Name_y']]

# Merge the datasets on 'countyFIPS' using the filtered columns
merged_datasets_final = pd.merge(df_quarter_A_filtered, super_team_dataset[columns_to_exclude], on='countyFIPS', how='inner')

merged_datasets_final.to_csv('merged_covid19_&_employment_data.csv', index=False)

print(merged_datasets_final)
```
