# Dataset Structure

## Columns Description

- **countyFIPS**: 
  - **Description**: The FIPS (Federal Information Processing Standards) code for the county. This is a numeric identifier unique to each county.
  - **Datatype**: Integer

- **County Name_x**: 
  - **Description**: The name of the county. This represents the county's name as of the dataset's recording.
  - **Datatype**: String (categorical)

- **State**: 
  - **Description**: The abbreviation of the state in which the county is located.
  - **Datatype**: String (categorical)

- **StateFIPS_cases**: 
  - **Description**: The FIPS code for the state, which is a numeric identifier for the state.
  - **Datatype**: Integer

- **2020-01-22_cases to 2023-07-23_cases**: 
  - **Description**: Columns representing the number of COVID-19 cases reported in each county on specific dates. Each column is named with the date in the format YYYY-MM-DD followed by "_cases".
  - **Datatype**: Integer

- **2020-01-22_deaths to 2023-07-23_deaths**: 
  - **Description**: Columns representing the number of COVID-19 deaths reported in each county on specific dates. Each column is named with the date in the format YYYY-MM-DD followed by "_deaths".
  - **Datatype**: Integer

- **County Name_y**: 
  - **Description**: This seems to be a duplicated column of County Name_x but might serve as a backup or for validation.
  - **Datatype**: String (categorical)

- **population**: 
  - **Description**: The population of the county. This provides demographic context for the COVID-19 case and death counts.
  - **Datatype**: Integer

### Understanding ACS Data Columns

Analyzing the columns in the dataset, `GEO_ID` and `Name` are self-explanatory. However, some columns like `DP03_0001E` and `DP03_0001M` might not be immediately clear. These columns refer to data elements from the American Community Survey (ACS) conducted by the U.S. Census Bureau. Specifically, they are part of the ACS 5-Year Estimates under the "Employment Status" section.

#### Breakdown of `DP03_0001E` and `DP03_0001M`

- **DP03**: The prefix indicates that the variable is part of the Data Profile (DP) series, specifically for the "Economic Characteristics" table.
  
- **0001**: This is the code for a particular data item within the Economic Characteristics section.

- **E**: The suffix "E" stands for "Estimate," indicating that this variable represents an estimate of the population characteristic.

- **M**: The suffix "M" stands for "Margin of Error," which provides a measure of the uncertainty or reliability of the estimate.

#### Detailed Description

- **`DP03_0001E` (Estimate)**: 
  - **Description**: Provides an estimate of the total number of people aged 16 years and older who are in the labor force. This includes those who are employed as well as those who are unemployed but actively seeking work.
  - **Usage**: This estimate helps in understanding the employment status of a population, including the size of the labor force in a given area.

- **`DP03_0001M` (Margin of Error)**:
  - **Description**: Provides the margin of error for the estimate, indicating the range within which the true value is likely to fall with a certain level of confidence (typically 90% confidence).
  - **Usage**: This margin of error is crucial for assessing the precision of the estimate and understanding the reliability of the data.

By breaking down these codes, we can better interpret the data and make more informed decisions based on the American Community Survey results.

### If you look closely at the data: 
In the COVID-19 data, the key fields are countyFIPS or a combination of County Name_x and State. In the enrichment data, the key is GEO_ID, which contains the FIPS code formatted as 0500000US followed by the FIPS code (e.g., 0500000US01001 for Autauga County). Clean the GEO_ID field by extracting the FIPS code from the GEO_ID column in the enrichment dataset by removing the 0500000US prefix. Then merge the datasets based on the common FIPS code.


```python
# This filters out rows where GEO_ID does not start with '0500000US'
df_clean_econ = df_clean_econ[df_clean_econ['GEO_ID'].str.startswith('0500000US')]
# Extract FIPS code from GEO_ID in enrichment data by removing the prefix '0500000US'
df_clean_econ['FIPS'] = df_clean_econ['GEO_ID'].str.replace('0500000US', '').astype(int)

# Merge the datasets on the FIPS code
merged_data = pd.merge(merged_df, df_clean_econ, left_on='countyFIPS', right_on='FIPS', how='left')
merged_data.head()
```

### Note:
There are quite a few NaN or null values in the “DP03_” columns. To combat this, we can simply replace the null values in these columns with the average value of the column that each of the null values reside in. This is arguably the safest approach in this case as dropping the column or row that these null values would result in the loss of tremendous amounts of data and thus, replacing these null values with the averages of their respective columns is the safest way of handling this. (Notice that we handle null values differently here than when we were handling null values in our enrichment data set where we dropped an entire column)

	


```python
merged_data = merged_data.apply(pd.to_numeric, errors='ignore')

# Calculate means of numeric columns
column_means = merged_data.mean(numeric_only=True)

# Replace NaN values with the mean of each column
merged_data_filled = merged_data.fillna(column_means)
pd.options.display.float_format = '{:.3f}'.format
# Verify if there are still any NaN values
print(merged_data_filled.isna().sum())
merged_data_filled.head()
```
