```python
import pandas as pd

# Load the dataset
super_covid19_data = pd.read_csv(r"C:\Users\bluev\Downloads\Group7ClassProject\Stage I\Team\super_covid19_data.csv")

# Filter columns for cases and deaths
cases_columns = [col for col in super_covid19_data.columns if '_cases' in col]
deaths_columns = [col for col in super_covid19_data.columns if '_deaths' in col]

# Include essential columns like countyFIPS, County Name, State, etc.
essential_columns = ['countyFIPS', 'County Name_x', 'State', 'StateFIPS_cases', 'population']

# Create separate DataFrames for cases and deaths
cases_df = super_covid19_data[essential_columns + cases_columns]
deaths_df = super_covid19_data[essential_columns + deaths_columns]

# Rename columns to remove '_cases' and '_deaths' suffix
cases_df.columns = [col.replace('_cases', '') for col in cases_df.columns]
deaths_df.columns = [col.replace('_deaths', '') for col in deaths_df.columns]

# Save the DataFrames into separate files
cases_df.to_csv(r"C:\Users\bluev\Downloads\Group7ClassProject\Stage I\Team\cases_data.csv", index=False)
deaths_df.to_csv(r"C:\Users\bluev\Downloads\Group7ClassProject\Stage I\Team\deaths_data.csv", index=False)
```
