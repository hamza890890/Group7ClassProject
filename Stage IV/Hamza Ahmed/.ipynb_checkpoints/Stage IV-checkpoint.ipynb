```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
import matplotlib.pyplot as plt

# Load data
enrichment_data = pd.read_csv(r"C:\Users\bluev\Downloads\5_states_and_CA.csv")
covid_data = pd.read_csv('US-Covid.csv')

# Filter for California and find the top 5 counties by total cases
covid_data_ca = covid_data[covid_data['State'] == 'CA']
# Use .loc to avoid SettingWithCopyWarning
covid_data_ca = covid_data_ca.copy()  # Create a copy of the DataFrame to work on
covid_data_ca['total_cases'] = covid_data_ca[[col for col in covid_data_ca.columns if '_cases' in col]].sum(axis=1)

top_counties = covid_data_ca.nlargest(5, 'total_cases')['County Name']

# Perform analysis for each top county
for county in top_counties:
    county_data = covid_data_ca[covid_data_ca['County Name'] == county]

    # Melt the DataFrame to long format
    county_data_long = county_data.melt(id_vars=['County Name', 'State', 'population'], 
                                         var_name='date_cases', 
                                         value_name='cases')
    
    # Extract date from the 'date_cases' column
    county_data_long['date'] = county_data_long['date_cases'].str.extract(r'(\d{4}-\d{2}-\d{2})')[0]
    county_data_long['date'] = pd.to_datetime(county_data_long['date'], errors='coerce')
    
    # Drop rows where date is NaT or cases are NaN
    county_data_long = county_data_long.dropna(subset=['date', 'cases'])
    
    # Set the date as index
    county_data_long = county_data_long.set_index('date')
    
    # Prepare regression model inputs
    county_cases = county_data_long['cases'].values
    days_since_start = np.arange(len(county_cases)).reshape(-1, 1)
    X = days_since_start
    y_cases = county_cases
    
    # Linear Regression
    linear_model = LinearRegression()
    linear_model.fit(X, y_cases)
    y_cases_pred_linear = linear_model.predict(X)
    
    # Polynomial Regression (Degree 3)
    poly = PolynomialFeatures(degree=3)
    X_poly = poly.fit_transform(X)
    poly_model = LinearRegression()
    poly_model.fit(X_poly, y_cases)
    y_cases_pred_poly = poly_model.predict(X_poly)
    
    # RMSE Calculation
    rmse_linear = np.sqrt(np.mean((y_cases - y_cases_pred_linear) ** 2))
    rmse_poly = np.sqrt(np.mean((y_cases - y_cases_pred_poly) ** 2))
    
    # Plot
    plt.figure(figsize=(10, 5))
    plt.plot(X, y_cases, label="Actual Cases")
    plt.plot(X, y_cases_pred_linear, label="Linear Prediction", linestyle="--")
    plt.plot(X, y_cases_pred_poly, label="Polynomial Prediction", linestyle="--")
    plt.title(f"{county} COVID-19 Cases Prediction")
    plt.xlabel("Days since 2020-06-01")
    plt.ylabel("New Cases")
    plt.legend()
    plt.show()
```

```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Load your datasets
enrichment_data = pd.read_csv(r"C:\Users\bluev\OneDrive\Documents\Economy.csv")
covid_data = pd.read_csv('US-Covid.csv')

# Normalize strings in the 'Label (Grouping)' column
enrichment_data['Label (Grouping)'] = enrichment_data['Label (Grouping)'].str.strip().str.lower()

# Function to get values safely and ensure they are numeric
def get_value(label):
    try:
        value = enrichment_data.loc[
            enrichment_data['Label (Grouping)'] == label.lower(),
            'California!!Estimate'
        ].values[0]
        return pd.to_numeric(value, errors='coerce')  # Convert to numeric
    except IndexError:
        print(f"Warning: No data found for '{label}'")
        return None

# Extract relevant values from the enrichment dataset
civilian_labor_force = get_value('Civilian labor force')
employed = get_value('Employed')
health_insurance_coverage = get_value('With health insurance coverage')

# Check if any of the values are None and handle accordingly
if civilian_labor_force is None or employed is None or health_insurance_coverage is None:
    print("One or more required values are missing. Please check the enrichment dataset.")
else:
    # Convert COVID-19 cases to numeric values
    covid_data.columns = covid_data.columns.str.replace('_cases', '')  # Remove '_cases' from column names for easier access
    covid_data.iloc[:, 5:] = covid_data.iloc[:, 5:].apply(pd.to_numeric, errors='coerce')  # Convert case columns to numeric

    # Hypothesis Testing Function
    def perform_hypothesis_test(covid_data, employment_value, covid_column):
        # Get cases per 10,000 population for the specified date
        covid_cases_per_10k = covid_data[covid_column] / (covid_data['population'] / 10000)

        # Create high and low groups based on employment value
        mean_cases = covid_cases_per_10k.mean()
        
        # Ensure employment_value is a float
        employment_value = float(employment_value)

        if employment_value > mean_cases:
            high_cases = covid_cases_per_10k[covid_cases_per_10k > mean_cases]
            low_cases = covid_cases_per_10k[covid_cases_per_10k <= mean_cases]
        else:
            high_cases = covid_cases_per_10k[covid_cases_per_10k <= mean_cases]
            low_cases = covid_cases_per_10k[covid_cases_per_10k > mean_cases]

        # Perform two-sample t-test
        t_stat, p_value = stats.ttest_ind(high_cases, low_cases, equal_var=False)  # Welch's t-test
        return t_stat, p_value

    # Perform hypothesis tests
    t_stat_civilian, p_value_civilian = perform_hypothesis_test(covid_data, civilian_labor_force, '2020-06-01')
    print(f'Civilian Labor Force - T-statistic: {t_stat_civilian}, P-value: {p_value_civilian}')

    t_stat_employed, p_value_employed = perform_hypothesis_test(covid_data, employed, '2020-06-01')
    print(f'Employed - T-statistic: {t_stat_employed}, P-value: {p_value_employed}')

    t_stat_health_insurance, p_value_health_insurance = perform_hypothesis_test(covid_data, health_insurance_coverage, '2020-06-01')
    print(f'Health Insurance Coverage - T-statistic: {t_stat_health_insurance}, P-value: {p_value_health_insurance}')

# Function to plot hypothesis results
def plot_hypothesis_results(covid_data, employment_value, covid_column, label):
    plt.figure(figsize=(12, 6))  # Increase figure size for better readability
    plt.title(f'COVID-19 Cases Based on {label}')

    # Calculate cases per 10,000
    covid_cases_per_10k = covid_data[covid_column] / (covid_data['population'] / 10000)

    # Drop NaN values to avoid plotting issues
    covid_cases_per_10k = covid_cases_per_10k.dropna()

    # Create histogram with more bins and adjusted range
    plt.hist(covid_cases_per_10k, bins=50, alpha=0.7, color='blue', label='COVID-19 Cases per 10,000')

    # Add a vertical line for employment value
    plt.axvline(employment_value, color='red', linestyle='dashed', linewidth=2, label='Employment Threshold')

    # Set x-axis limits (adjust these values based on your data range)
    plt.xlim(0, covid_cases_per_10k.max() + 5)  # Add a little padding

    # Adding grid lines for better readability
    plt.grid(axis='y', linestyle='--', alpha=0.7)

    # Set labels and legend
    plt.xlabel('Cases per 10,000')
    plt.ylabel('Frequency')
    plt.legend()
    plt.tight_layout()  # Automatically adjust subplot parameters to give specified padding
    plt.show()

# Plot results for each hypothesis with appropriate labels
plot_hypothesis_results(covid_data, civilian_labor_force, '2020-06-01', 'Civilian Labor Force')
plot_hypothesis_results(covid_data, employed, '2020-06-01', 'Employed')
plot_hypothesis_results(covid_data, health_insurance_coverage, '2020-06-01', 'Health Insurance Coverage')
```

```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Load your datasets
enrichment_data = pd.read_csv(r"C:\Users\bluev\OneDrive\Documents\Economy.csv")
covid_data = pd.read_csv('US-Covid.csv')

# Normalize strings in the 'Label (Grouping)' column
enrichment_data['Label (Grouping)'] = enrichment_data['Label (Grouping)'].str.strip().str.lower()

# Function to get values safely and ensure they are numeric
def get_value(label):
    try:
        value = enrichment_data.loc[
            enrichment_data['Label (Grouping)'] == label.lower(),
            'California!!Estimate'
        ].values[0]
        return pd.to_numeric(value, errors='coerce')  # Convert to numeric
    except IndexError:
        print(f"Warning: No data found for '{label}'")
        return None

# Extract relevant values from the enrichment dataset
civilian_labor_force = get_value('Civilian labor force')
employed = get_value('Employed')
health_insurance_coverage = get_value('With health insurance coverage')

# Check if any of the values are None and handle accordingly
if civilian_labor_force is None or employed is None or health_insurance_coverage is None:
    print("One or more required values are missing. Please check the enrichment dataset.")
else:
    # Convert COVID-19 cases to numeric values
    covid_data.columns = covid_data.columns.str.replace('_cases', '')  # Remove '_cases' from column names for easier access
    covid_data.iloc[:, 5:] = covid_data.iloc[:, 5:].apply(pd.to_numeric, errors='coerce')  # Convert case columns to numeric
    print(covid_data.head())  # Show a sample of the data
    print(covid_data.isna().sum())  # Check for NaN values

    # Define the date range
    start_date = '2020-06-01'
    end_date = '2021-01-10'
    date_range = pd.date_range(start=start_date, end=end_date)

    # Aggregate data
    aggregated_cases = pd.Series(index=covid_data.index, dtype=float).fillna(0)  # Initialize with 0

    for date in date_range:
        date_str = date.strftime('%Y-%m-%d')  # Convert date to string format
        if date_str in covid_data.columns:  # Check if the date column exists
            # Add to aggregated cases, ignoring NaN values
            aggregated_cases += covid_data[date_str].fillna(0)

    # Display the result
    print("Aggregated Cases after filling NaN:\n", aggregated_cases)

    # Calculate cases per 10,000 population for the aggregated data
    total_population = covid_data['population'].sum()
    if total_population > 0:  # Ensure total population is not zero
        cases_per_10k = aggregated_cases / (total_population / 10000)
    else:
        print("Total population is zero or invalid. Check the population column.")

    # Drop NaN values from cases_per_10k before hypothesis testing and plotting
    cases_per_10k = cases_per_10k.dropna()
    print("Cases per 10k:\n", cases_per_10k)

    # Hypothesis Testing Function
    def perform_hypothesis_test(cases_per_10k, employment_value):
        # Create high and low groups based on employment value
        mean_cases = cases_per_10k.mean()
        print(f"Mean cases: {mean_cases}")

        # Ensure employment_value is a float
        employment_value = float(employment_value)

        if employment_value > mean_cases:
            high_cases = cases_per_10k[cases_per_10k > mean_cases]
            low_cases = cases_per_10k[cases_per_10k <= mean_cases]
        else:
            high_cases = cases_per_10k[cases_per_10k <= mean_cases]
            low_cases = cases_per_10k[cases_per_10k > mean_cases]

        print(f"High cases count: {len(high_cases)}, Low cases count: {len(low_cases)}")

        # Perform two-sample t-test
        if len(high_cases) > 0 and len(low_cases) > 0:
            t_stat, p_value = stats.ttest_ind(high_cases, low_cases, equal_var=False)  # Welch's t-test
            return t_stat, p_value
        else:
            print("Warning: Not enough data to perform hypothesis test.")
            return np.nan, np.nan

    # Perform hypothesis tests
    t_stat_civilian, p_value_civilian = perform_hypothesis_test(cases_per_10k, civilian_labor_force)
    print(f'Overall - Civilian Labor Force - T-statistic: {t_stat_civilian}, P-value: {p_value_civilian}')

    t_stat_employed, p_value_employed = perform_hypothesis_test(cases_per_10k, employed)
    print(f'Overall - Employed - T-statistic: {t_stat_employed}, P-value: {p_value_employed}')

    t_stat_health_insurance, p_value_health_insurance = perform_hypothesis_test(cases_per_10k, health_insurance_coverage)
    print(f'Overall - Health Insurance Coverage - T-statistic: {t_stat_health_insurance}, P-value: {p_value_health_insurance}')

    # Function to plot hypothesis results
    def plot_hypothesis_results(cases_per_10k, employment_value, label):
        if cases_per_10k.empty:
            print(f"No data available to plot for {label}.")
            return
        
        plt.figure(figsize=(12, 6))  # Increase figure size for better readability
        plt.title(f'COVID-19 Cases Based on {label}')

        # Create histogram of aggregated cases per 10,000
        plt.hist(cases_per_10k, bins=50, alpha=0.7, color='blue', label='COVID-19 Cases per 10,000')

        # Add a vertical line for employment value
        plt.axvline(employment_value, color='red', linestyle='dashed', linewidth=2, label='Employment Threshold')

        # Set x-axis limits
        plt.xlim(0, cases_per_10k.max() + 5)  # Add a little padding

        # Adding grid lines for better readability
        plt.grid(axis='y', linestyle='--', alpha=0.7)

        # Set labels and legend
        plt.xlabel('Cases per 10,000')
        plt.ylabel('Frequency')
        plt.legend()
        plt.show()

    # Plot results
    plot_hypothesis_results(cases_per_10k, civilian_labor_force, 'Civilian Labor Force')
    plot_hypothesis_results(cases_per_10k, employed, 'Employed')
    plot_hypothesis_results(cases_per_10k, health_insurance_coverage, 'Health Insurance Coverage')
```
