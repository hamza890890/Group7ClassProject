```python
import pandas as pd

cases_df = pd.read_csv('covid_confirmed_usafacts.csv')
deaths_df = pd.read_csv('covid_deaths_usafacts.csv')
population_df = pd.read_csv('covid_county_population_usafacts.csv')


merged_cases_deaths = pd.merge(cases_df, deaths_df, on=['countyFIPS', 'County Name', 'State'], suffixes=('_cases', '_deaths'))


super_covid_df = pd.merge(merged_cases_deaths, population_df, on=['countyFIPS', 'State'])

merged_df = super_covid_df.drop_duplicates()


print(merged_df)

merged_df.to_csv('super_covid19_data.csv', index=False)

```
